## 📁 Dataset Description
This dataset contains personal expense records including date, category, and amount spent.
It is used to analyze spending behavior and financial habits.

```python
data = pd.DataFrame({
    "Date": pd.date_range(start="2024-01-01", periods=10),
    "Category": ["Food","Travel","Shopping","Food","Bills","Entertainment","Food","Savings","Travel","Shopping"],
    "Amount": [180,120,600,150,300,250,200,500,100,400]
})

data
```

```python
total_expense = data["Amount"].sum()

category_analysis = data.groupby("Category")["Amount"].sum().sort_values(ascending=False)

print("💰 Total Expense:", total_expense)
print("\n📊 Category-wise Spending:\n", category_analysis)
```

## 📊 Insights Generated
- Identifies highest spending category
- Tracks total financial usage
- Shows spending behavior patterns
- Helps improve savings decisions

```python
essential = data[data["Category"].isin(["Food", "Bills", "Savings"])]["Amount"].sum()

score = round((essential / total_expense) * 100, 2)

print("🏦 Financial Discipline Score:", score, "%")
```

```python
def ai_suggestion(row):
    if row["Category"] == "Shopping" and row["Amount"] > 500:
        return "⚠ High shopping detected - avoid impulse buying"
    elif row["Category"] == "Food" and row["Amount"] > 200:
        return "⚠ Food expense high - try home meals"
    elif row["Category"] == "Savings":
        return "✅ Excellent savings habit"
    else:
        return "✔ Normal spending"

data["AI_Insight"] = data.apply(ai_suggestion, axis=1)

data
```

```python
category_analysis.plot(kind="bar")
plt.title("📊 Expense Category Analysis")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.show()
```
